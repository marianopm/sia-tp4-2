```python
import os
import pandas as pd

data_europe = pd.read_csv('./data/europe.csv', skiprows=1, header=None)
data_europe_with_column_names = pd.read_csv('./data/europe.csv')
valores_primera_columna = data_europe_with_column_names['Country'].values

# Mostrar los valores de la primera columna
print(valores_primera_columna)


# Le quito la primer columna
data_europe_with_column_names = data_europe_with_column_names.iloc[:, 1:]

# Le quito la primer columna
data_europe = data_europe.iloc[:, 1:]

print(data_europe.head())
print(data_europe_with_column_names.head())
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns


print(data_europe)
# boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(data=pd.DataFrame(data_europe))
for y in range(-4, 4, 1):  # Ajusta el rango y el paso según tus necesidades
    plt.axhline(y, color='gray', linestyle='--', linewidth=0.5)
plt.title('Box Plot')
plt.ylabel('Principal Component')
plt.xlabel('Country characteristics')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Estandariza la data
scaler = StandardScaler()
standarized_data = scaler.fit_transform(data_europe)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns


print(data_europe)
# boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(data=pd.DataFrame(standarized_data, columns=data_europe_with_column_names.columns))
for y in range(-4, 4, 1):  # Ajusta el rango y el paso según tus necesidades
    plt.axhline(y, color='gray', linestyle='--', linewidth=0.5)
plt.title('Box Plot')
plt.ylabel('Principal Component')
plt.xlabel('Country characteristics')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlación
corr_matrix = pd.DataFrame(standarized_data).corr()

# Crear el heatmap de la matriz de correlación
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black', 
                      xticklabels=data_europe_with_column_names.columns, yticklabels=data_europe_with_column_names.columns)
plt.title('Matriz de Correlación')
plt.show()
```

Aqui los colores saturados indican una alta correlación  y los colores suaves indican lo contrario.

```python
# Se reduce a 2 componentes principales
pca = PCA(n_components=2)
pca_values = pca.fit_transform(standarized_data)

# Agrego los valores obtenidos
data_europe['PC1'] = pca_values[:, 0]
data_europe['PC2'] = pca_values[:, 1]

# Obtener las cargas de los componentes principales
loadings = pca.components_.T

# Convertir las cargas a un DataFrame para mejor legibilidad
loadings_df = pd.DataFrame(loadings, columns=[f'PC{i+1}' for i in range(loadings.shape[1])], index=data_europe_with_column_names.columns)

print(loadings_df)
```

```python
import numpy as np

# Boxplot
print( data_europe['PC1'])
print(valores_primera_columna)
plt.figure(figsize=(10, 8))
bars = plt.bar(valores_primera_columna, data_europe['PC1'])

for bar in bars:
    if bar.get_height() < 0:
        bar.set_color('red')
        
for y in range(-5, 5, 1):  # Ajusta el rango y el paso según tus necesidades
    plt.axhline(y, color='gray', linestyle='--', linewidth=0.5)
plt.xticks(rotation=90)
plt.title('PCA1 per Country')
plt.xlabel('Country')
plt.ylabel('PCA1')
plt.show()
```

Para PCA1 > 0, el país tiene mas inflación, Area, desempleo y nivel militar.
Para PCA1 < 0, el país tiene mas expectativa de vida, crecimiento poblacional y PBI 

```python
# Crear el biplot
df = data_europe
df.columns =  ['Area', 'GDP', 'Inflation', 'Life.expect', 'Military', 'Pop.growth', 'Unemployment', 'PC1', 'PC2']

# Crear el biplot
plt.figure(figsize=(10, 8))
for i in range(len(df)):
    plt.scatter(df['PC1'].iloc[i], df['PC2'].iloc[i], color='blue', s=5)  # Scatter plot de PC1 vs PC2
    plt.text(df['PC1'].iloc[i], df['PC2'].iloc[i], valores_primera_columna[i], fontsize=7, alpha=0.8)  # Etiquetas de los países

print(df.columns[:-2])
# Agregar flechas para cada característica
for i, feature in enumerate(df.columns[:-2]):  # Iterar sobre las características
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], color='red', alpha=0.5, head_width=0.05, head_length=0.05)
    plt.text(pca.components_[0, i], pca.components_[1, i]*1.2, feature, color='red', fontsize=7)

# Ajustar los límites del eje
plt.xlim(df['PC1'].min() - 0.1, df['PC1'].max() + 0.1)
plt.ylim(df['PC2'].min() - 0.1, df['PC2'].max() + 0.1)

# Añadir título y etiquetas de los ejes
plt.title('Biplot de PC1 vs PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Mostrar el biplot
plt.grid(True)
plt.show()
```

```python
# Crear el biplot
df = data_europe

df.columns =  ['Area', 'GDP', 'Inflation', 'Life.expect', 'Military', 'Pop.growth', 'Unemployment', 'PC1', 'PC2']

escala = 1.5 # Factor de escala
pca_components_scaled = pca.components_ * escala

# Crear el biplot
plt.figure(figsize=(10, 8))
for i in range(len(df)):
    plt.scatter(df['PC1'].iloc[i], df['PC2'].iloc[i], color='blue')  # Scatter plot de PC1 vs PC2
    plt.text(df['PC1'].iloc[i], df['PC2'].iloc[i], valores_primera_columna[i], fontsize=8)  # Etiquetas de los países

# Agregar flechas para cada característica
for i, feature in enumerate(df.columns[:-2]):  # Iterar sobre las características
    plt.arrow(0, 0, pca_components_scaled[0, i], pca_components_scaled[1, i], color='red', alpha=0.5)
    plt.text(pca_components_scaled[0, i], pca_components_scaled[1, i], feature, color='red', fontsize=10)

# Ajustar los límites del eje
plt.xlim(df['PC1'].min() - 1, df['PC1'].max() + 1)
plt.ylim(df['PC2'].min() - 1, df['PC2'].max() + 1)

# Añadir título y etiquetas de los ejes
plt.title('Biplot PC1 vs PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Mostrar el biplot
plt.grid(True)
plt.show()
```
